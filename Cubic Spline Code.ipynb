```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return np.exp(-x)*np.cos(10*x**2)

def quadf(a,b,c,x):
    return a*x**2 + b*x + c

### INITIALIZATION ###############

# NUMBER OF ENTRIES IN X_VEC (0,n-1)
n = 6#33 # N>=3

# CREATE AN ARRAY OF X VALUES
#xVec = np.sqrt(np.arange(0,n)/8) #xVec =np.linspace(0,2,n)
xVec = np.array([0,10,15,20,22.5,30])

# CREATE AN ARRAY OF Y VALUES
#fVec = f(xVec)
fVec = np.array([0,227.04,362.78,517.35,602.97,901.67])


# CREATE AN ARRAY AS A 4(N-1) x 4(N-1) MATRIX OF ZEROS
Amx = np.zeros((4*(n-1),4*(n-1)))

# CREATE THE VECTOR OF CONSTANTS AS A 3(N-1) x 1 COLUMN VECTOR OF ZEROS
bVec = np.zeros((4*(n-1),1))

### CONTRUCTION OF MATRIX ########


k = 0 # USED FOR COLUMN INDEX for 'Amx'
xIndex = 0 # USED TO SELECT X-VALUE for 'xVec' and 'fVec'



for ii in range(0,2*(n-1)): # INPUT ENTRIES ROWS 0,..,2*(n-1) FOR 'Amx' AND 'bVec'
    
    if(ii % 2 == 1):
        xIndex+=1

    if(ii % 2 ==0 and not(ii==0) ):
        k+=1
        
    bVec[ii] = fVec[xIndex]
    
    Amx[ii,4*k] = xVec[xIndex]**3
    Amx[ii,4*k+1] = xVec[xIndex]**2
    Amx[ii,4*k+2] = xVec[xIndex]
    Amx[ii,4*k+3] = 1
    

        

cur = 0
for kk in range(2*(n-1),3*(n-1)-1): # INPUT ENTRIES ROWS 2(N-1),...,3(N-1)-1
    
    Amx[kk,4*cur] = 3*xVec[cur+1]**2
    Amx[kk,4*cur+1] = 2*xVec[cur+1]
    Amx[kk,4*cur+2] = 1
    
    Amx[kk,4*(cur+1)] = -3*xVec[cur+1]**2
    Amx[kk,4*(cur+1) + 1] = -2*xVec[cur+1]
    Amx[kk,4*(cur+1) + 2] = -1
    
    cur+=1

cur = 0 
for kk in range(3*(n-1)-1,4*(n-4)):  # INPUT ENTRIES ROWS 3(N-1)-1,...,4(N-1)-2
    
    
    Amx[kk,6*cur] = 6*xVec[cur+1]
    Amx[kk,6*cur+1] = 1
    
    
    Amx[kk,6*(cur+1)-2] = -6*xVec[cur+1]
    Amx[kk,6*(cur+1) -1] = -1
    
    cur+=1    
    
    
    
    
    
### Populate last rows ####   
    
Amx[-1,0]=1 # iNPUT ENTRIES FOR LAST ROW
Amx[-2,-1]=1
 
text_file = open("Output.txt", "w")
text_file.write(np.array_str(Amx,max_line_width=150))
text_file.close()  
    
coeff = np.linalg.solve(Amx,bVec) # RREF



# ### Visualization ##############

# for jj in range(n-1):
    
#     index = 3*jj
    
#     # CREATE AN ARRAY OF EVENLY SPACED VALUES FROM 'xi' to 'xi+1'
#     x_values = np.linspace(xVec[jj],xVec[jj+1],100) 

#     if(jj==0):
#         y_values = coeff[1]*x_values + coeff[2]
#         plt.plot(x_values,y_values, 'b')
#     else:   
        
#         y_values = quadf(coeff[index],coeff[index+1],coeff[index+2],x_values)
#         plt.plot(x_values,y_values,'b')

# ## Labels ##

# plt.xlabel('x')
# plt.ylabel('y')
# plt.title('Spline Method for function: \n $e^{-x}*cos(10x^2)$')

# #plt.savefig('Spline')
# plt.show()

# ### AREA UNDER THE CURVE ################

# summa = 0

# frac = np.array([1/3,1/2,1])
# exp = np.array([3,2,1])

# for kk in range(n-1):
#     index=3*kk
    
#     if(kk == 0):
        
#         x_values = np.array([xVec[kk+1]]*2)**exp[1:] - np.array([xVec[kk]]*2)**exp[1:]
#         # INTEGRAL OF A LINEAR EQUATION EVALUATED FROM 'x0' to 'x1' 
#         summa+= np.sum(frac[1:] * x_values * coeff[index+1:index+3].flatten())   
        
#     else:
#         x_values = np.array([xVec[kk+1]]*3)**exp - np.array([xVec[kk]]*3)**exp
#         # INTEGRAL OF A QUADRATIC EQUATION EVALUATED FROM 'xI' to 'xi+1' 
#         summa+= np.sum(frac * x_values * coeff[index:index+3].flatten())  
        
        
        
        
# print("Est. Integral for Function from zero to two:",summa)
        
```
