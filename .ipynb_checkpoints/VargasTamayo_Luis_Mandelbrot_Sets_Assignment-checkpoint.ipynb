### Question 8

For 100 evenly spaced values of θ from 0 to 2π, plot the values of c(θ)  in the complex plane. (in your plot, you should be plotting the real part of c on the x axis and the imaginary part of c on the y axis).  Draw an important conclusion from your result.

$c(\theta) =\frac{1}{4} - \frac{1}{4}[ cis(\theta)-1 ]^2$ <br><br>
$c(\theta) =\frac{1}{4} - \frac{1}{4}[ cos(\theta)+i*sin(\theta)-1 ]^2$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def c(theta):
    return 0.25 - 0.25 * (np.cos(theta) + 1j*np.sin(theta)-1)**2

theta = np.linspace(0,2*np.pi,100)

cVal = c(theta)

plt.plot(cVal.real,cVal.imag,"*")


# These are the stable fixed points in the Mandelbrot set.
# This looks exacly like the outline of the Cardoid Shape from the Mandelbrot Graph.
# The period is one
```

### Question 10

Take 16 points on  the circle |c+1| = 0.25. Create 3 plots: on the first plot, show 30 terms of the Mandelbrot sequence for the 16 points.  For the second plot use ‘figure(2)’ in Python), show only  the even terms in the Mandelbrot sequence (i.e. $z_0$,$z_2$,$z_4$,…).  On the third  plot, show only the odd terms of the Mandelbrot sequence (i.e. $z_1,z_3,z_5,$…) .  Repeat for points on the circles |c+1| = 0.2  and $|c+1| = 0.1$. Explain what your plots are showing about the behavior of the Mandelbrot sequence for points within the circle  |c+1| = 0.25.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 

####################################################
# n: number of iterations for set
# m: number of different points (c) in the graph
def program(n,m,r,hs):
    
    def Mandelbrot(x, c):
        return x**2 + c
    
    theta = np.arange(m)*(2*np.pi/m) # Array with different thetas
    cVec= r*(np.cos(theta)+1j*np.sin(theta)) +hs  # Array with different points

    values=np.zeros((n,m))*1j

    temp = cVec

    for xx in range(n):
        values[xx] = temp
        temp= Mandelbrot(temp, cVec)
        
    return values
################################################### 
    
even = np.arange(0,30,2)
odd = np.arange(1,30,2)
other = np.arange(2,30,3)    


for ii in range(1,4):
    fig = plt.figure(ii)
    

    
    if(ii==1):
        fig.suptitle('$|c+1| = 0.25$', fontsize=16)
        values = program(n=30, m=16, r=0.25, hs=-1)    
    elif(ii==2):
        fig.suptitle('$|c+1| = 0.2$', fontsize=16)
        values = program(n=30, m=16, r=0.2, hs=-1)
    elif(ii==3):
        fig.suptitle('$|c+1| = 0.1$', fontsize=16)
        values = program(n=30, m=16, r=0.1, hs=-1)
    
    
    ax1 = fig.add_subplot(1,3,1)
    ax1.plot(values.real,values.imag,'.') 
    ax1.set_title("Maldelbrot Set for three different 'c' \n with 16 elements on the set")
    ax1.set_xlabel('Real Part')
    ax1.set_ylabel('Imag Part')

    ax2 = fig.add_subplot(1,3,2)
    ax2.plot(values.real[even],values.imag[even],'.') 
    ax2.set_title("Maldelbrot Set for three different 'c' \n with even elements from the set")
    ax2.set_xlabel('Real Part')
    ax2.set_ylabel('Imag Part')

    ax3 = fig.add_subplot(1,3,3)
    ax3.plot(values.real[odd],values.imag[odd],'.')  
    ax3.set_title("Maldelbrot Set for three different 'c' \n with odd elements from the set")
    ax3.set_xlabel('Real Part')
    ax3.set_ylabel('Imag Part')

    
    fig.subplots_adjust(right=1.5, top = 0.75,wspace=0.8)


# Given the equation |c+1| = 0.25, we are looking at the bulb to the left of the cardoid. 
# Looking at equation |c+1| = 0.1, we are only looking at half the 0.16 of the area from |c+1| = 0.25.
# We see that as the area of observation increases, more points from the area are conversing to a
# to a single point. We also see that their are two groups in the graph. This tells us that
# the period for the bulb is two.
```

### Question 11:

Use the reference to locate a bulb  (also called “decoration”)  of period 3. Locate 3 points inside the bulb, and make 4 plots.   on the first plot, show all terms of the Mandelbrot sequence for the 3 points.  For the second plot, show the terms (i.e. z_0,z_3,z_6,…).  On the third  plot, show only the terms  z_1,z_4,z_7,… . On the fourth plot, show the terms z_2,z_5,z_8,… . Explain how your results show that the sequence is period 3.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 

n = 16 # length of maldebrot set
m = 3 # Number of sets of Maldebrot (number of c points)

cVec = np.array( [-0.06+0.74j, -0.12+0.75j, -0.18+0.76j])

def Mandelbrot(x, c):
    return x**2 + c

values= np.zeros((n,m))*1j

temp = cVec

for xx in range(n):
    values[xx] = temp
    temp= Mandelbrot(temp, cVec)
    
real_lst = values.real
imag_lst = values.imag

even = np.arange(0,n,3)
odd = np.arange(1,n,3)
other = np.arange(2,n,3)
    
plt.figure(1)
plt.plot(real_lst, imag_lst,'.')   
plt.title("Maldelbrot Set for three different 'c' \n with 16 elements on the set")
plt.xlabel('Real Part')
plt.ylabel('Imag part')

plt.figure(2)
plt.plot(real_lst[even], imag_lst[even],'*')
plt.title("Maldelbrot Set for three different 'c' \n with {0,3,6,..} elements from the set")
plt.xlabel('Real Part')
plt.ylabel('Imag part')

plt.figure(3)
plt.plot(real_lst[odd], imag_lst[odd],'*')
plt.title("Maldelbrot Set for three different 'c' \n with {1,4,7,...} elements from the set")
plt.xlabel('Real Part')
plt.ylabel('Imag part')

plt.figure(4)
plt.plot(real_lst[other], imag_lst[other],'*')
plt.title("Maldelbrot Set for three different 'c' \n with {2,5,8,...} elements from the set")
plt.xlabel('Real Part')
plt.ylabel('Imag part')

# After finding three different points on a bulb that contained a period of 3, we created a 
# maldelbrot set for each one. Then, we graphed the points.
# At first glance, there are three different clusters with three groups composed of 
# three different clusters.
# when we saw the even elements from the set, we still saw the three different clusters with 
# three different groups with three 
# different clusters.
# when we saw the odd elements from the set, we still saw the three different clusters with 
# three different groups with three 
# different clusters.
# when we saw the {2,5,..} elements from the set, we only saw the three different groups with three 
# different clusters.
# This tells us that the period of the bulb is three. The values of the mandelbrot set for each 'c'
# are moving from cluster to cluster.
```

### Question 12:

Another set that is closely related to the Mandelbrot set is the Julia set. The Julia set for the complex number c  is the set of complex numbers z such that the iterated function sequence ${z,f_c (z),f_c(f_c (z)),f_c (f_c (f_c (z))),…,}$ remains bounded.  Note the difference between Julia sets and Mandelbrot sets: for the Mandelbrot set the starting point is always c ,while for the Julia set the starting point can be any complex number z. <br><br>
As with the Mandelbrot set, we can make color plots using the following function:
Index(z) = the smallest value of n such that $|z_n > 2|$, where z_n is the n’th term of the iterated function sequence. 
Take your Mandelbrot set code and modify it to create a diaram of the Julia set corresponding to c=-0.52+0.57i, which is shown below.<br><br>

In your plot, take $-2≤x,y≤2$  (note that x and y are the real and imaginary parts respectively of the complex number z). At the beginning, you may take 200 x values and 200 y values, and take the maximum number of iterations equal to 32.  Your colors may not be the same as in the picture, but the shape should be the same.


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


### PART A #######################

x_min = -2 # Cant Changed
x_max = 2
y_min = -2
y_max = 2 # cant changed
index_max = 32

n_x = 200
n_y = 200

c = -0.52+0.57j

xVals = np.linspace(x_min,x_max,n_x) # (START,END,QUANTITY)
yVals = np.linspace(y_min,y_max,n_y)  # (START,END,QUANTITY)
indexVals = np.zeros((n_x,n_y)) # (ROWS,COLUMNS)


### PART B #######################

for ii in range(n_x):
    for jj in range(n_y):
        
        z = xVals[ii] + yVals[jj]*1j #r
        
        indexVals[ii,jj] = index_max
        
        for kk in range(index_max):

            z = z**2 + c
            
            if(np.abs(z)>2):
                indexVals[ii,jj]= kk+1
                break


plt.figure(1)
plt.title('Segment of boundary of Mandelbrot set')
plt.ylabel('y')
plt.xlabel('x')
    
plt.imshow(indexVals,cmap='jet')
plt.colorbar()  
plt.gca().invert_yaxis()

```
