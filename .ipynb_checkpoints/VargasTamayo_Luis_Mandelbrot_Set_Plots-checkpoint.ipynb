### Part A: Initialization

Your code should begin with an “initialization” section where all numbers are represented as parameters. This is where you put all the numbers in your code – there should be no numbers at all in the computation section of your code. 

### Part B: Calculations

You  will need a for ii in range(n_r)  loop to range through the θvalues, another for jj in range(n_theta) loop inside to range through the r values, and an innermost for kk in range(indexMax)  loop to evaluate Index for the given r and θ  values.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


### PART A #######################

r_min = 0.25
r_max = 0.35
theta_min = np.pi/3 #(2 * np.pi) / 3
theta_max = np.pi /2 #(5 * np.pi) / 6
index_max = 64

n_r = 450
n_theta = 500


rVals = np.linspace(r_min,r_max,n_r) # (START,END,QUANTITY)
thVals = np.linspace(theta_min,theta_max,n_theta)  # (START,END,QUANTITY)
indexVals = np.zeros((n_r,n_theta)) # (ROWS,COLUMNS)


### PART B #######################

for ii in range(n_r):
    for jj in range(n_theta):
        
        c = -1 + rVals[ii]*np.cos(thVals[jj]) + 1j*rVals[ii]*np.sin(thVals[jj])
        z = c
        
        indexVals[ii,jj] = index_max
        
        for kk in range(index_max):
            # f(x)=x^2 + c
            # f(c) = c^2 + c = z
            z = z**2 + c
            
            if(np.abs(z)>2):
                indexVals[ii,jj]= kk+1
                break



c = cmap = plt.cm.get_cmap("winter")

plt.figure(1)
plt.title('Segment of boundary of Mandelbrot set')
plt.ylabel('Theta')
plt.xlabel('Radius')
    
plt.imshow(indexVals,cmap='jet')
plt.colorbar()  
plt.gca().invert_yaxis()


plt.figure(2)
plt.title('Segment of boundary of Mandelbrot set \n(Zoomed in to x:[25,75] y:[0,45])')
plt.ylabel('Theta')
plt.xlabel('Radius')
    
plt.imshow(indexVals,cmap='jet')
plt.xlim([25,75])
plt.ylim([0,45])
plt.colorbar() 

```
