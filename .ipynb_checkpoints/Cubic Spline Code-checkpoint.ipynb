## Cubic Spline Method: Natural Spline

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return np.exp(-x)*np.cos(10*x**2)

def cubicf(a,b,c,d,x):
    return a*x**3 + b*x**2 + c*x + d


### INITIALIZATION ###############

# NUMBER OF ENTRIES IN X_VEC (0,n-1)
n = 32 # N>3
#n = 5

# CREATE AN ARRAY OF X VALUES
xVec = np.sqrt(np.arange(0,n+1)/8)
#xVec = np.array([0,10,15,20,22.5,30])
#xVec = np.array([1,2,3])

# CREATE AN ARRAY OF Y VALUES
fVec = f(xVec)
#fVec = np.array([0,227.04,362.78,517.35,602.97,901.67])
#fVec = np.array([1,5,4])


# CREATE AN ARRAY AS A 4(N-1) x 4(N-1) MATRIX OF ZEROS
Amx = np.zeros((4*n,4*n))

# CREATE THE VECTOR OF CONSTANTS AS A 3(N-1) x 1 COLUMN VECTOR OF ZEROS
bVec = np.zeros((4*n,1))

################################
### CONTRUCTION OF MATRIX ######
################################

# FILL ROWS 0,2,...,2N
for ii in range(n):
    row = 2*ii
    col = 4*ii
    
    Amx[row,col:col+4] = xVec[ii]**3, xVec[ii]**2, xVec[ii], 1
    bVec[row] = fVec[ii]

# FILL ROWS 1,3,...,2N
for ii in range(n):
    row = 2*ii + 1
    col = 4*ii
    
    Amx[row,col:(col+4)] = xVec[ii+1]**3, xVec[ii+1]**2, xVec[ii+1], 1
    bVec[row] = fVec[ii+1]

# FILL ROWS 2N,2N+1,...,3N-1
for ii in range(n-1):
    row = 2*n + ii
    col = 4*ii
    col2 = col + 4
    
    Amx[row,col:(col+3)] = 3*xVec[ii+1]**2,2*xVec[ii+1], 1
    Amx[row,col2:(col2+3)] = -3*xVec[ii+1]**2,-2*xVec[ii+1], -1
    
    

# FILL ROWS 3N-1,3N,...,4N-1
for ii in range(1,n):
    row = 3*n +(-2+ii)
    col = 4*(ii-1)
    col2 = col + 4
    
    Amx[row,col:(col+2)] = 6*xVec[ii], 2
    Amx[row,col2:(col2+2)] = -6*xVec[ii], -2
    
    
    
# FILL ROW 3N
Amx[-2,0:2] = 6 * xVec[0], 2
Amx[-1,-4:-2] = 6 * xVec[-1], 2

#print(Amx)

# SOLUTIONS
coeff = np.linalg.solve(Amx,bVec)



################################
### VISUALIZATION ##############
################################

for jj in range(n):
    
    index = 4*jj
    
    # CREATE AN ARRAY OF EVENLY SPACED VALUES FROM 'xi' to 'xi+1'
    x_values = np.linspace(xVec[jj],xVec[jj+1],100) 

    y_values = cubicf(*coeff[index:index+4],x_values)
    plt.plot(x_values,y_values,'b')

## Labels ##

plt.xlabel('x')
plt.ylabel('y')
plt.title('Spline Method for function: \n $e^{-x}*cos(10x^2)$ \n Natural Spline')

#plt.savefig('Natural')
plt.show()

################################
### AREA UNDER THE CURVE #######
################################

summa = 0

frac = np.array([1/4,1/3,1/2,1])
exp = np.array([4,3,2,1])

for kk in range(n):
    index=4*kk
    
    x_values = np.array([xVec[kk+1]]*4)**exp - np.array([xVec[kk]]*4)**exp

    # INTEGRAL OF A QUADRATIC EQUATION EVALUATED FROM 'xi' to 'xi+1' 
    summa+= np.sum(frac * x_values * coeff[index:index+4].flatten())  

print("Est. Integral for Function from zero to two:",summa)
```
