## Taylor Series Assignment

The $N$th order Taylor series formula is:<br>
$$ f(x+h) \approx f(x) + \frac{h^1}{1!} f^{(1)}(x) +  \frac{h^2}{2!} f^{(2)}(x) + \ldots + \frac{h^N}{N!} f^{(N)}(x),$$

where $f^{(j)}(x)$ denotes the $j$th derivative of the function $f$ at the point $x$. or<br>

$$f(x) \approx f(a) + \frac{(x-a)^1}{1!} f^{(1)}(a) +  \frac{(x-a)^2}{2!} f^{(2)}(a) + \ldots + \frac{(x-a)^N}{N!} f^{(N)}(a)$$

where $f^{(j)}(a)$ denotes the $j$th derivative of the function $f$ at the point $a$.



<br>The error in the formula will be less than the Taylor error bound, which is given by:

$$ R_N(x+h) = \frac{|h|^{N+1}}{(N+1)!}\max_{y \in [x,x+h]}| f^{(N+1)}(y)|.$$  or <br>

$$ R_N(x) = \frac{|x-a|^{N+1}}{(N+1)!}\max_{y \in [x-a,x]}| f^{(N+1)}(y)|.$$

```python
#Warm-up exercise: Complete the following code that computes and plots 
# the 7th order Taylor series approximation and Taylor error bound for 
# f(x) = cos(x) with expansion point x=0 on the interval [-3, 3] 
# with increment 0.01.  
#
### Import section
# 3 necessary statements for numerical plotting in Jupyter
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# factorial comes from 'math' package
from math import factorial



# define function
def fun(x):
    return np.cos(x)

#
### Initialize section
#array of derivatives at 0 compute by hand
coefs = np.array([1,0,-1,0,1,0,-1,0])

# degree of Taylor polynomial
dg = 7

# Max abs. value of N+1 deriv on interval (Compute using calculus)
# (used in error formula:  error <= abs(h^(N+1)/(N+1)) * max(abs(N+1th deriv.))
maxD = 1.0

#array with h values
hVec = np.arange(-3,3.01,0.01)
#hVec = np.linspace(-3.0,3.0,0.01)

# Give the expansion point for the Taylor series
x = 0
# Create an array to contain Taylor series
taylor = np.zeros(len(hVec))
#
### Calculation section
#calculate Taylor series for all h at once using a loop

for jj in range(len(coefs)):
    taylor += (hVec**jj)*coefs[jj]/factorial(jj)
    

#actual function values 
truVals = fun(hVec+x)

# absolute error actual and taylor series
errVec = np.abs(truVals - taylor)

#theoretical error=|h^(deg+1)/(deg+1)!*max|n+1 deriv||
theoErr = (np.abs((hVec**(dg+1))/factorial(dg+1)) * maxD)

#Compute actual x values (displaced by expansion point)
xVals = x + hVec

#
### Output section
#plot figures for cos and errors using
# Object-oriented style plotting
fig = plt.figure()

# Generate a grid of 1x2 subplots and get
# axes objects for 1st location
ax1 = fig.add_subplot(1,2,1)
ax1.set_title('Taylor Series and Actual Function')

# axes objects for 2st location
ax2 = fig.add_subplot(1,2,2)
ax2.set_title('Actual and Theo. error')



# objects for the left subplot ( attached to axs #0)
ax1.plot(xVals,taylor,label='Taylor Series')
ax1.plot(xVals,truVals,label='Actual Function')


# objects for the right subplot (attached to axes #1)
ax2.plot(xVals,theoErr,label='Theoretical Error')
ax2.plot(xVals,errVec,label='Calculated Error')

ax1.legend()
ax2.legend()

plt.show()
```
