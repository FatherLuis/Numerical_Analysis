### Question 8

For 100 evenly spaced values of θ from 0 to 2π, plot the values of c(θ)  in the complex plane. (in your plot, you should be plotting the real part of c on the x axis and the imaginary part of c on the y axis).  Draw an important conclusion from your result.

$c(\theta) =\frac{1}{4} - \frac{1}{4}[ cis(\theta)-1 ]^2$ <br><br>
$c(\theta) =\frac{1}{4} - \frac{1}{4}[ cos(\theta)+i*sin(\theta)-1 ]^2$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def c(theta):
    return 0.25 - 0.25 * (np.cos(theta) + 1j*np.sin(theta)-1)**2

theta = np.linspace(0,2*np.pi,100)

cVal = c(theta)

plt.plot(cVal.real,cVal.imag)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 

n = 30 #umber of different points on the graph
m = 100 # Number of different Thetas
r = 0.25
theta = np.arange(m)*(2*np.pi/m) # Array with different thetas
cVec= r*(np.cos(theta)+1j*np.sin(theta))  # Array with different points

def Mandelbrot(x, c):
    return x**2 + c

values=np.zeros((n,m))*1j

temp = cVec

for xx in range(n):
    values[xx] = temp
    temp= Mandelbrot(temp, cVec)

    
plt.figure(1)
plt.plot(values.real,values.imag,'.')    
    
    

plt.figure(2)
plt.plot([x for x in range(n) if x%2==0],
         [x for x in range(n) if x%2==0],'.')    

plt.figure(3)
plt.plot([x for x in range(n) if x%2==1],
         [x for x in range(n) if x%2==1],'.')  
```

### Question 11:

Use the reference to locate a bulb  (also called “decoration”)  of period 3. Locate 3 points inside the bulb, and make 4 plots.   on the first plot, show all terms of the Mandelbrot sequence for the 3 points.  For the second plot, show the terms (i.e. z_0,z_3,z_6,…).  On the third  plot, show only the terms  z_1,z_4,z_7,… . On the fourth plot, show the terms z_2,z_5,z_8,… . Explain how your results show that the sequence is period 3.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 

n = 16 # length of maldebrot set
m = 3 # Number of sets of Maldebrot

cVec=np.array( [-0.06+0.74j, -0.12+0.75j, -0.18+0.76j])

def Mandelbrot(x, c):
    return x**2 + c

values= np.zeros((n,m))*1j

temp = cVec

for xx in range(n):
    values[xx] = temp
    temp= Mandelbrot(temp, cVec)
    
real_lst = values.real
imag_lst = values.imag

even = np.arange(0,n,2)
odd = np.arange(1,n,2)
other = np.arange(2,n,3)
    
plt.figure(1)
plt.plot(real_lst, imag_lst,'.')   
plt.title("Maldelbrot Set for three different 'c' \n with 16 elements on the set")
plt.xlabel('Real Part')
plt.ylabel('Imag part')

plt.figure(2)
plt.plot(real_lst[even], imag_lst[even],'*')
plt.title("Maldelbrot Set for three different 'c' \n with even elements from the set")
plt.xlabel('Real Part')
plt.ylabel('Imag part')

plt.figure(3)
plt.plot(real_lst[odd], imag_lst[odd],'*')
plt.title("Maldelbrot Set for three different 'c' \n with odd elements from the set")
plt.xlabel('Real Part')
plt.ylabel('Imag part')

plt.figure(4)
plt.plot(real_lst[other], imag_lst[other],'*')
plt.title("Maldelbrot Set for three different 'c' \n with (2,5,8,...) elements from the set")
plt.xlabel('Real Part')
plt.ylabel('Imag part')
```
