```python
# The following code finds the forward, backward, and centered difference
# derivative approximations for a given function for a range of delta 
# values, then plots the absolute errors on a log-log scale plot.

# "Magic" statement for jupyter notebook to display plots
%matplotlib inline

# Necessary imports
import numpy as np
import matplotlib.pyplot as plt


## Initialization
# Point where derivative is evaluated
x= np.pi / 4

#function to be differentiated
def f(x):
    return np.sin(x)**2

#Derivative of function (theo. calculation)
def df(x):
    return 2*np.sin(x)*np.cos(x)

#get vector of deltax values
dx = np.array([3*10**x for x in range(-1,-10,-1)])


## Calculation
#Compute forward approx.
forwardApprox = (f(x+dx) - f(x))/dx

#Compute backward approx
backwardApprox = (f(x) - f(x-dx))/dx

#use average of forward and backward for centered
centeredApprox = (f(x+dx) - f(x-dx))/(2*dx)


#compute absolute errors
trueVal = df(x)

abs_err_forward = np.abs(trueVal-forwardApprox)
abs_err_backward = np.abs(trueVal-backwardApprox)
abs_err_centered = np.abs(trueVal-centeredApprox)


## Display section -- create figure

#plot forward, backward, and centered errors on log-log scale
plt.loglog(dx,abs_err_forward,label='Forward Diff')
plt.loglog(dx,abs_err_backward,label='Backward Diff')
plt.loglog(dx,abs_err_centered,label='Centered Diff')

plt.gca().invert_xaxis()





# If limits are specified, set plot limits



# Use log-log scale


# Label plot and axes
plt.xlabel('$\Delta x$')
plt.ylabel('Error')


# Add legend and title
plt.title('Three Types of Difference\n Derivatives Approximation')
plt.legend()
```
